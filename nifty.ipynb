```python
ticker = "^NSEI"
start_date = "2020-01-01"
end_date = "2023-12-31"

df_stock = yf.download(ticker, start=start_date, end=end_date)
df_stock.reset_index(inplace=True)
df_stock.rename(columns={'Date': 'date'}, inplace=True)
df_stock.head()
```

```python
import pandas as pd

# Load all headline files
cnbc = pd.read_csv('cnbc_headlines.csv')
guardian = pd.read_csv('guardian_headlines.csv')
reuters = pd.read_csv('reuters_headlines.csv')

# Standardize column names (adjust based on actual columns in your CSV files)
# Example: If columns are "date" and "headline"
cnbc = cnbc[['date', 'headline']]
guardian = guardian[['date', 'headline']]
reuters = reuters[['date', 'headline']]

# Combine all headlines into one DataFrame
all_news = pd.concat([cnbc, guardian, reuters], ignore_index=True)

# Convert date column to datetime
all_news['date'] = pd.to_datetime(all_news['date']).dt.date
all_news['date'] = pd.to_datetime(all_news['date'])

# Sort by date
all_news.sort_values('date', inplace=True)
all_news.head()
```
